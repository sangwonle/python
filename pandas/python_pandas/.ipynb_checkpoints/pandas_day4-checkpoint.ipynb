```python
# %load ../font.py
#!/usr/bin/env python

# In[ ]:
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

import platform

if platform.system() == 'Darwin': #맥
        plt.rc('font', family='AppleGothic') 
elif platform.system() == 'Windows': #윈도우
        plt.rc('font', family='Malgun Gothic') 
elif platform.system() == 'Linux': #리눅스 (구글 콜랩)
        plt.rc('font', family='Malgun Gothic') 

plt.rcParams['axes.unicode_minus'] = False #한글 폰트 사용시 마이너스 폰트 깨짐 해결

```

```python
titanic = sns.load_dataset('titanic')
print(titanic.head())
print(titanic.info())
```

```python
# 회귀선이 있는 scatter 그래프, 회귀선이 없는 2 개의 그래프
sns.set_style('darkgrid')
fig = plt.figure(figsize=(15,6))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)


sns.regplot(x='age', y='fare', data=titanic, ax=ax1)
sns.regplot(x='age', y='fare', data=titanic, ax=ax2, fit_reg=False)

plt.show()
```

```python
# distplot() : 커널 밀도
# 스타일 테마 설정 (darkgrid)

sns.set_style('darkgrid')

# 그래프 객체 생성 (figure에 3개의 서브 플롯을 생성)
fig = plt.figure(figsize=(15, 5))
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)
 
# distplot
sns.distplot(x=titanic.fare, ax=ax1) 

# kdeplot
sns.kdeplot(x='fare', data=titanic, ax=ax2) 

# histplot
sns.histplot(x='fare', data=titanic,  ax=ax3)

# 차트 제목 표시
ax1.set_title('titanic fare - distplot')
ax2.set_title('titanic fare - kedplot')
ax3.set_title('titanic fare - histplot')

plt.show()
```

```python
# heatmap

sns.set_style('darkgrid')

table = titanic.pivot_table(index=['sex'], columns=['class'], aggfunc='size')


sns.heatmap(table, annot=True, fmt='d', cmap='YlGnBu', linewidth=0.5, cbar=True)

plt.show()
```

```python
import json

# 경기도 인구변화 데이터를 불러와서 데이터프레임으로 변환
file_path = './dataset/경기도인구데이터.xlsx'
df = pd.read_excel(file_path, index_col='구분', engine= 'openpyxl')
df.columns = df.columns.map(str)

# 경기도 시군구 경계 정보를 가진 geo-json 파일 불러오기
geo_path = './dataset/경기도행정구역경계.json'
try:
    geo_data = json.load(open(geo_path, encoding='utf-8'))
except:
    geo_data = json.load(open(geo_path, encoding='utf-8-sig'))

# 경기도 지도 만들기
g_map = folium.Map(location=[37.5502,126.982], 
                   tiles='Stamen Terrain', zoom_start=9)

# 출력할 연도 선택 (2007 ~ 2017년 중에서 선택)
year = '2017'

# Choropleth 클래스로 단계구분도 표시하기
folium.Choropleth(geo_data=geo_data,    # 지도 경계
                 data = df[year],      # 표시하려는 데이터
                 columns = [df.index, df[year]],  # 열 지정
                 fill_color='YlOrRd', fill_opacity=0.7, line_opacity=0.3,
                 threshold_scale=[10000, 100000, 300000, 500000, 700000],
                 key_on='feature.properties.name',
                 ).add_to(g_map)

# 지도를 HTML 파일로 저장하기
g_map.save('./dataset/gyonggipopulation' + year + '.html')


```

```python
# 누락데이터 처리

df = sns.load_dataset('titanic')
df.info()
```

```python
# deck 컬럼의 값들의 갯수를 세어 봄 df.value.count() 유니크한 값들의 갯수
print(df['deck'].value_counts())
print()
deck_cnt = df['deck'].value_counts(dropna=False) # Nan 데이터 계산하기
print(deck_cnt)
print()

# isnull() 메서드로 누락 된 데이터 찾기, null 이면 True, 값이 있으면 False 반환
print(df['deck'].head().isnull())

# isnull()과 sum() 함수를 이용하여 누락 된 데이터 갯수 찾기
df['deck'].isnull().sum()
df.isnull().sum(axis=0) # age, deck, embarked, embark_town => null data 포함

```

```python
# 반복무으로 Nan 데이터 갯수 계산하기

nan_data = df.isnull()
nan_data
for col in nan_data.columns:
    nan_cnt = nan_data[col].value_counts() # 각 열의 Nan 갯수 파악
    try:
        print(col, ':', nan_cnt[True]) #값이 있으면 출력
    
    except:
        print(col, ':', 0)
print()

```

```python
# Nan 데이터 확인 -> 어떻게 처리 할건지
# Nan 데이터가 있는 컬럼을 삭제, 혹은 Nan 데이터가 있는 행을 삭제 할껀지

# df.dropna() : axis=1 열삭제 axis=0 행을 삭제
# dropna() :  Nan이 존재하는 컬럼을 삭제, 갯수가 500 이상인 컬러만 삭제

df_thresh = df.dropna(axis=1, thresh=500)
# df_thresh.info()

print( )
# age에 Nan이 존재하는 행을 삭제
df_age = df.dropna(subset=['age'], how='any', axis=0)
# df_age.info()


# 컬럼 리스트 : Nan이 존재하는 컬럼리스트
nan_col = [df.isnull().sum() > 0]
nan_col_names = list(df.columns[nan_col])
# Nan이 존재하는 모든 행을 삭제
df_nan = df.dropna(subset=nan_col_names, how='any', axis=0)
print(df_nan.info())
```

```python
# 누락 데이터 치환 : df.fillna(값 또는 method=ffill / bfill, inplace=True)

print(df['age'].isnull().sum()) # Nan 존재함

# Nan이 입력 된 age를 평균 나이로 지환
df_age = df.copy()
df_age['age'].fillna(df_age['age'].mean(axis=0), inplace=True)
df_age['age'].isnull().sum()
print(df['age'].head(10), df_age['age'].head(10))


```

```python
# 누락 데이터 치환
# embark_town 825~ 830

df['embark_town'][825:832]
df['embark_town'].value_counts().idxmax() # 가장 빈번하게 발생하는 인덱스 명
df_em = df.copy()
df_em['embark_town'].fillna(df_em['embark_town'].value_counts().idxmax(),
                           inplace=True)
df_em['embark_town'][825:832]

```

```python
# 누락 데이터 치환 : 이전 데이터로 치환
df_me = df.copy()
df_me['embark_town'].fillna(method='ffill', inplace=True)
print(df_me.embark_town[[828, 829]], df.embark_town[[828,829]])
```

```python
# 중복 데이터 처리 : df.duplicated() -> 중복 엽 확인
df1 = pd.DataFrame({'c1': ['a', 'a', 'b', 'a', 'b'],
                   'c2': [1,1,1,2,2],
                    'c3': [1,1,2,2,2]})
print(df1)
df_dup = df1.duplicated()
col_dup = df1['c1'].duplicated()
print(col_dup)

# 중복된 데이터를 제거 행을 제거
df2 = df1.drop_duplicates()
df2

# 중복된 컬럼을 제거
df3 = df1.drop_duplicates(subset=['c2', 'c3'])
df3
```

```python
# titanic에서 age, fare, class, alive 컬럼 가져와서 df_titanic로 저장 후
# Nan 있는 컬럼의 값의 숫자는 평균으로 문자는 이전 값으로 대체
# 중복 된 행과 컬럼은 삭제

df_titanic =  df.loc[: ,['age', 'fare', 'class', 'alive']]
df_titanic.head()


for col in df_titanic.columns:
    if df_titanic[col].dtype == 'float64':
        df_titanic[col].fillna(df_titanic[col].mean(), inplace=True)
    else:
        df_titanic[col].fillna(method='ffill', inplace=True)
df_titanic.head(6)

len(df_titanic)

df_titanic.drop_duplicates(inplace=True)
len(df_titanic)

df_titanic.drop_duplicates(subset=['age'], inplace=True)
len(df_titanic)
```

```python
# dataset/auto_mpg.csv
# 데이터 표준화

df = pd.read_csv('dataset/auto-mpg.csv', header=None)

df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']

df.info()
df.head()
```

```python
# 단위 환산 : mpg > gallon > kg으로 변환

mpg_to_kg = 1.60934 / 3.78541

df['kpl'] = (df['mpg'] * mpg_to_kg).round(2)
df.head(2)
```

```python
# 자료형 변환 : object -> float으로 변경
df['horsepower'].unique() # ? 발견 > Nan으로 변경후 Nan 처리함수 사용
df['horsepower'].replace('?', np.nan, inplace=True)
df['horsepower'].unique()

# 데이터 타입을 float으로 변경
df['horsepower'] = df['horsepower'].astype('float')
```

```python
df.info()
```

```python
# origin
df.origin.unique() # 숫자를 카테고리 형태로 변환

df.origin.replace({1: 'USA', 2: 'EU', 3: 'KOR'}, inplace=True)
df.origin.unique()

# object를 카테고리로 변경

df.origin = df.origin.astype('category')
df.info()
```

```python
# 제조 년도 : model year -> 카테고리 변경
df['model year'].unique()

# 연도 숫자로 category 변경
df['model year'] = df['model year'].astype('category')
df.info()
```

```python
# 데이터 전처리
# 1. 누락 데이터 처리 : 제거, 치환
# 2. 중복 데이터 처리 : 제거(행), 컬럼...
# 3. 자료 표준화
#    단위 환산, 자료형 변경 -> object - float, object - category
# 변화 불가능한 자료는 Nan으로 치환하고 자료형 변경
```
